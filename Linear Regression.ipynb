```python
#Importing Modules
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
#Creating Dataset
np.random.seed(20)
x, y = make_regression(n_samples = 200,
                     n_features = 1,
                      noise = 10)
```

```python
#Splitting Dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
#Model
class LinearRegression:
    def __init__(self, epochs = 1000, lr = 0.001):
        self.epochs = epochs
        self.lr = lr
    def predict(self, x_test):
            return np.dot(x_test, self.weigh) + self.bias
            
    def plot(self, x_test, y_pred):
        plt.scatter(x_test, 
                    y_pred)
        plt.plot(x_test, self.predict(x_test), c = "orange")
        
    def fit(self, x_train, y_train):
        rows, cols = x_train.shape
        self.weigh = np.zeros((cols, 1))
        self.bias = 1
        
        plt.scatter(x_train, y_train)
        
        for i in range (self.epochs):
            
            #Calculating Predicting Y
            y = np.dot(x_train, self.weigh) + self.bias
            
            #Minimizing the Cost
            jPar_w = 1/rows * (y.reshape((rows,)) - y_train)
            jPar_w =  np.dot(jPar_w, x_train)
                               
            jPar_b = 1/rows * (y.reshape((rows,)) - y_train)
            
            #Summation
            jPar_w = jPar_w.sum()
            jPar_b = jPar_b.sum()

            self.weigh = self.weigh - (self.lr * jPar_w)
            self.bias = self.bias - (self.lr * jPar_b)
            
            if(i % 500 == 0):
                plt.plot(x_train, self.predict(x_train))
            
        return self
            
        
```

```python
Model = LinearRegression(10000)
```

```python
Model.fit(x_train, y_train);
```

```python
Model.predict(x_test);
```

```python
Model.plot(x_test, y_test)
```
